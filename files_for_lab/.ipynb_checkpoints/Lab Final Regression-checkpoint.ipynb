```python
#reading the data
target = pd.read_csv("target.csv")
numerical = pd.read_csv("numerical.csv")
categorical = pd.read_csv("categorical.csv")
```

```python
#Full dataframe
full_df = pd.concat([categorical, numerical, target], axis=1)
```

```python
#creating a updated full dataframe with only the rows where target _B is 1(only donors)
full_df = pd.DataFrame(full_df[full_df["TARGET_B"] == 1])
```

```python
#splitting cat and num
numerical = full_df.select_dtypes(include=["float64", "int64"])
categorical = full_df.select_dtypes(exclude=["float64", "int64"])
```

```python
target = numerical[["TARGET_D"]]
```

```python
numerical = numerical.drop("TARGET_D", axis=1)
numerical = numerical.drop("TARGET_B", axis=1)
```

```python
#now we have our target_D, numerical without target and our categoricals.
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
numerical_scaled = scaler.fit_transform(numerical)

numerical_scaled = pd.DataFrame(numerical_scaled, columns=numerical.columns)
```

```python
X = numerical_scaled
y = target["TARGET_D"]

from sklearn.feature_selection import RFE
from sklearn import linear_model
lm = linear_model.LinearRegression()
rfe = RFE(lm, n_features_to_select=30, verbose=False)
rfe.fit(X, y)
```

```python
df = pd.DataFrame(data = rfe.ranking_, columns=["Rank"])
df["Column_name"] = pd.DataFrame(numerical).columns
df[df["Rank"]==1]

selected_columns = df[df["Rank"]==1]

keep_list = list(selected_columns["Column_name"])
keep_list
```
